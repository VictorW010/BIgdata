```python
from pyspark.sql.functions import udf
# create the sample array.
target_trend = ['increase','increase','increase','decrease','decrease','decrease',\
                'decrease','increase','increase','increase','decrease','decrease']
# pyspark.sql.functions.lit(col) Creates a Column of literal value.
target_trend = F.array([F.lit(x) for x in target_trend])
```

```python
# Create the UDF function to distingush the trend
def To_trend(x):
    trend = []
    for i in range(len(x) - 1):
        diff = x[i+1] - x[i]
        if diff > 0:
            trend.append('increase')
        elif diff < 0:
            trend.append('decrease')
    return trend
#Register the function as UDF
To_trend = udf(To_trend, ArrayType(StringType()))
```

3.1.2 The reasons behind the performance difference between the RDD job and Dataframe job

Answer:
By comparing with the two cases above. the conclusion is obviously that the Dataframe job is faster than the RDD.

As far as i am concerned, firstly, RDD tends to create new objects rather than modify old ones in most situations. It makes the program tend to create a large number of temporary objects at run time, putting pressure on processors. Dataframes can constantly reuse previous variables to reduce object allocation. Secondly, because i am willing to check the situation of the metadate before programing, I found that, compared with the RDD job, the Dataframe job has more reasonable and explicit data structure information, namely schema. Consequently, the Dataframes has higher execution efficiency and shorter execution time.
