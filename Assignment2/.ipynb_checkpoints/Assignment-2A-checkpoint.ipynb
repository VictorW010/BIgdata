2.1.2 Write code to create the columns based on your discussion above

```python
# convert 'Day'
df_join = df_join.withColumn('Day', F.dayofweek("Date_Time"))
# convert 'Month'
df_join = df_join.withColumn('Month', F.month("Date_Time"))
```

```python
# select appropriate columns
features = ['Year', 'Month', 'Mdate', 'Day','Time','Sensor_ID', 'direction_1','direction_2','latitude', 'longitude', 'status']
```

```python
# prediction column
df_join_nextHour = df_join.withColumn('Time', df_join.Time + 1).withColumnRenamed('Hourly_counts', 'Preivous_Hour_Counts')\
                    .select('Time', 'Preivous_Hour_Counts', 'Year', 'Month', 'Mdate', 'Sensor_ID')
#between 9:00am and midnight.
df_join_nextHour = df_join.join(df_join_nextHour, ['Time', 'Year', 'Month', 'Mdate', 'Sensor_ID'])\
                     .filter(F.col('Time') >= 9)\
                     .na.fill('Null').cache()
```

```python
features.append('Preivous_Hour_Counts')
```

```python
df_join_features = df_join_nextHour.select(features)

df_join_features.show(2, truncate = False)
```

2.2 Preparing Spark ML Transformers/Estimators for features, labels and models

2.2.1 Write code to create Transformers/Estimators for transforming/assembling the columns you selected above in 2.1, and create ML model Estimators for Decision Tree and Gradient Boosted Tree model for each use case

```python
df_join_features.printSchema()
```

```python
# cheack the numerical column
numerical_joinF = [x.name for x in df_join_features.schema.fields if (isinstance(x.dataType, IntegerType) or isinstance(x.dataType, DoubleType))]
numerical_joinF
```

```python
non_numerical_joinF = [x for x in df_join_features.columns if x not in numerical_joinF]
non_numerical_joinF 
```

```python
#create Transformers/Estimators for transforming/assembling the columns
#Initialize StringIndexer (use inputCols and outputCols)
inputCols = non_numerical_joinF 
outputCols = [f'{x}_index' for x in non_numerical_joinF ]
stage_1 = StringIndexer(inputCols=inputCols, outputCols=outputCols)

#Define OneHotEncoder with the appropriate columns
inputCols_OHE = [f'{x}_index' for x in non_numerical_joinF]
outputCols_OHE = [f'{x}_vec' for x in inputCols]
stage_2 = OneHotEncoder(inputCols=inputCols_OHE, outputCols=outputCols_OHE)

#Define the assembler with appropriate input and output columns
inputCols = outputCols_OHE + numerical_joinF
stage_3 = VectorAssembler(inputCols=inputCols, outputCol='features')

#create ML model Estimators for Decision Tree 
#Extracts the number of nodes in the decision tree and the tree depth in the model and stores it in dt.
classifier_dt = DecisionTreeClassifier(featuresCol = 'features', labelCol = 'above_threshold', maxDepth = 5)
#create Gradient Boosted Tree model for each use case
classifier_gbt = GBTClassifier(featuresCol='features', labelCol='above_threshold')
```

2.2.2 Write code to include the above Transformers/Estimators into pipelines

```python
pipeline_classifier_dt  = Pipeline(stages=[stage_1, stage_2, stage_3, classifier_dt])
pipeline_classifier_gbt = Pipeline(stages=[stage_1, stage_2, stage_3, classifier_gbt])
```

Compute the AUC, accuracy, recall and precision for the above-threshold/below-threshold label from each model testing result using pyspark MLlib/ML APIs

```python
prediction_dt.groupBy('above_threshold', 'prediction').count().show()
```

```python
prediction_gbt.groupBy('above_threshold', 'prediction').count().show()
```

```python
def compute_metrics(predictions):
    #WRITE CODE: to calculate accuracy,precision,recall and f1 based on above example
    TN = predictions.filter('prediction = 0 AND above_threshold= 0').count()
    TP = predictions.filter('prediction = 1 AND above_threshold = 1').count()
    FN = predictions.filter('prediction = 0 AND above_threshold = 1').count()
    FP = predictions.filter('prediction = 1 AND above_threshold = 0').count()
    accuracy = (TP + TN)/(TN+TP+FN +FP)
    precision = TP/(TP+FN)
    recall = TP/(TP+FP)
    f1= 2/(1/recall + 1/precision)
    
    return accuracy,precision,recall,f1    
```

```python
predction_dict=[]
predction_dict.append(compute_metrics(prediction_dt))
predction_dict.append(compute_metrics(prediction_gbt))
# predction_dict
for x in range(0,len(predction_dict)):
        print('accuracy: %.8f, recall:%.8f, precision:%.8f, f1:%.8f'% predction_dict[x])
```

```python
# display the difference between Models (from Lab 6)
import matplotlib.pyplot as plt
    #matplotlib inline
def plot_metrics(x,y):
    plt.style.use('ggplot')   
    x_pos = [i for i, _ in enumerate(x)]
    plt.bar(x_pos, y, color='blue')
    plt.xlabel("Classification Algorithms")
    plt.ylabel("AUC")
    plt.title("Accuracy of ML Classification Algorithms")
    plt.xticks(x_pos, x)
    plt.show()
    return 
```

```python
for x in range(0,len(predction_dict)):
    plot_metrics(["accuracy","precision","recall","f1"], predction_dict[x])
```
